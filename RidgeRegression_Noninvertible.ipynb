```python
import numpy as np
import plotly.graph_objects as go

X = np.array([[1, 2],
              [2, 1],
              [3, 5],
              [4, 3]])
Y = np.array([5, 9, 13, 17])

#add an intercept term for ridge regression
X_with_intercept = np.hstack([np.ones((X.shape[0], 1)), X])

#ridge regression setup
lamda_ridge = 0.1
I = np.eye(X_with_intercept.shape[1])
beta_ridge = np.linalg.inv(X_with_intercept.T @ X_with_intercept + lamda_ridge * I) @ X_with_intercept.T @ Y

#create the grid for plotting
x1_range = np.linspace(0,5,20)
X2_range = 2*x1_range
X1_mesh, X2_mesh = np.meshgrid(x1_range, X2_range)
Y_mesh = beta_ridge[0] + beta_ridge[1]*X1_mesh + beta_ridge[2]*X2_mesh

#plotting using plotly
fig = go.Figure(data=[go.Scatter3d(x=X[:, 0], y=X[:, 1], z=Y, mode='markers', marker=dict(size=5, color='red')),
                     go.Surface(x=X1_mesh, y=X2_mesh, z=Y_mesh, opacity=0.5)])
fig.update_layout(title='Ridge Regression Surface Plot',
                  scene=dict(xaxis_title='X1', yaxis_title='X2', zaxis_title='Y'),
                  width=800, height=800)
fig.show()
```
